```python
import os
import numpy as np
from statistics import stdev
root_dir = 'results-2021'
os.chdir('..')
```

```python
def type_result(syn):
    ##single hop
    #type 1 2 3
    for i in sorted(syn.keys()):
        if len(i.split('_')) != 3:
            continue
        if 'r50' in i:
            continue

        print(i, *syn[i])
        
    #type 4
    for i in sorted(syn.keys()):
        if len(i.split('_')) != 3:
            continue
        if 'r50' in i and 'uu' in i:
            print(i, *syn[i])
    
    ##multi hop
    #type 1 2 3
    for i in sorted(syn.keys()):
        if len(i.split('_')) != 4:
            continue
        if 'r50' in i or '_40' in i:
            continue

        print(i, *syn[i])
        
    #type 4
    for i in sorted(syn.keys()):
        if len(i.split('_')) != 4:
            continue
        if 'r50' in i and 'uu' in i and '_40' not in i:
            print(i, *syn[i])
            
    #type 5
    for i in sorted(syn.keys()):
        if len(i.split('_')) != 4:
            continue
        if 'r25' in i and 'uu' in i and '_40' in i:
            print(i, *syn[i])
```

# MFEA 2

```python
res = {}

for fn in os.listdir(os.path.join(root_dir, 'mfea11')):
    lines = open(os.path.join(root_dir, 'mfea11', fn)).readlines()
    tests = lines[0].replace(',', '').replace('[', '').replace(']', '').replace("'", '')
    if tests in res.keys():
        res[tests].append([float(tmp.replace('\n', '')) for tmp in lines[-2:]])
    else:
        res[tests] = [[float(tmp.replace('\n', '')) for tmp in lines[-2:]]]
        
    values = [float(tmp.replace('\n', '')) for tmp in lines[-2:]]
        
    if any([tmp > 10 for tmp in values]):
        print(fn)
        cnt+=1
```

```python
syn = {}

for k in sorted(res.keys()):
    tasks = k.split(' ')
    for i in range(len(tasks)):
        print(tasks[i].split('/')[-1].split('.')[0], end=' ')
        values = [tmp[i] for tmp in res[k]]
        if any([tmp == float('inf') for tmp in values]):
            for j in range(len(values)):
                if values[j] == float('inf'):
                    values[j] = min(values) 
        print(min(values), np.mean(values), stdev(values), end=' ')
        v = tasks[i].split('/')[-1].split('.')[0]
        if v in syn.keys():
            if syn[v][1] > np.mean(values):
                syn[v] = [min(values), np.mean(values), stdev(values)]  
        else:
            syn[v] = [min(values), np.mean(values), stdev(values)]
    print()
```

```python
type_result(syn)
```
