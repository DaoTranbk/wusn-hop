```python
import os
import numpy as np
from statistics import stdev
```

```python
single = {}
multi = {}
single_time = {}
multi_time = {}
for fn in os.listdir('mfea4'):
    lines = open(os.path.join('mfea4', fn)).readlines()
    tests = lines[0].replace(',', '').replace('[', '').replace(']', '').replace("'", '')
    start = int(lines[1].split(' ')[-1])
    end = int(lines[-5].split(' ')[-1])
    if len(lines) != 501:
        continue
#     try:
    if len(fn.split('_')) == 5:
        if tests in multi.keys():
            multi[tests].append([float(tmp.replace('\n', '')) for tmp in lines[-4:]])
        else:
            multi[tests] = [[float(tmp.replace('\n', '')) for tmp in lines[-4:]]]
            
        if tests in multi_time.keys():
            multi_time[tests].append(end - start)
        else:
            multi_time[tests] = [end - start]

    else:
        if tests in single.keys():
            single[tests].append([float(tmp.replace('\n', '')) for tmp in lines[-4:]])
        else:
            single[tests] = [[float(tmp.replace('\n', '')) for tmp in lines[-4:]]]
        
        if tests in single_time.keys():
            single_time[tests].append(end-start)
        else:
            single_time[tests] = [end-start]
            
#     except:
#         print(fn, lines[-4:])
```
