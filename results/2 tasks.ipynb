```python
import os
import numpy as np

hop_data = {}

for fn in os.listdir('hop'):
    with open(os.path.join('hop', fn)) as f:
        lines = [tmp.replace('\n', '') for tmp in f.readlines()]
        hop_file = lines[0]
        res = float(lines[-1].split('\t')[-1])
        if hop_file not in hop_data.keys():
            hop_data[hop_file] = {fn: res}
        else:
            hop_data[hop_file][fn] = res
        
layer_data = {}

for fn in os.listdir('layer'):
    with open(os.path.join('layer', fn)) as f:
        lines = [tmp.replace('\n', '') for tmp in f.readlines()]
        hop_file = lines[0]
        res = float(lines[-1].split('\t')[-1])
        if hop_file not in layer_data.keys():
            layer_data[hop_file] = {fn: res}
        else:
            layer_data[hop_file][fn] = res
print(len(hop_data), len(layer_data))
```

```python
import os
import numpy as np

hop_time = {}


for fn in os.listdir('hop'):
    with open(os.path.join('hop', fn)) as f:
        lines = [tmp.replace('\n', '') for tmp in f.readlines()]
        hop_file = lines[0]
        start = int(lines[1].split(' ')[-1])
        end = int(lines[-2].split(' ')[-1])
        if hop_file in hop_time.keys():
            hop_time[hop_file].append(end-start)
        else:
            hop_time[hop_file] = [end-start]
        
layer_time = {}

for fn in os.listdir('layer'):
    with open(os.path.join('layer', fn)) as f:
        lines = [tmp.replace('\n', '') for tmp in f.readlines()]
        hop_file = lines[0]
        start = int(lines[1].split(' ')[-1])
        end = int(lines[-2].split(' ')[-1])
        if hop_file in layer_time.keys():
            layer_time[hop_file].append(end-start)
        else:
            layer_time[hop_file] = [end-start]
```

```python
for i in sorted(layer_time.keys()):
    print(i, min(layer_time[i]), np.mean(layer_time[i]))
    
for i in sorted(hop_time.keys()):
    print(i, min(hop_time[i]), np.mean(hop_time[i]))
```
