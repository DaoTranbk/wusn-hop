```python
import os
from utils.input import WusnInput, SensorNode, RelayNode, Point
from utils.point import distance
from queue import Queue
```

```python
def max_hop(wusnInput):
    depth = {}

    q = Queue()
    depth[wusnInput.BS] = 0
    q.put(wusnInput.BS)
    
    for rl in wusnInput.relays:
        q.put(rl)
        depth[rl] = 1
    while q.qsize() != 0:
        u = q.get()
        for ss in wusnInput.sensors:
            if distance(ss, u) <= 2 * wusnInput.radius:
                if ss not in depth or depth[u] + 1 < depth[ss]:
                    depth[ss] = depth[u] + 1
                    q.put(ss)
    print(list(depth.values()))
    return max(list(depth.values()))
    
max_depths = []
for fn in os.listdir('data/large/hop'):
    if fn.startswith('.'):
        continue
#     print(fn)
    wusnInput = WusnInput.from_file(os.path.join('data/large/hop', fn))
    max_depths.append(max_hop(wusnInput))
```
