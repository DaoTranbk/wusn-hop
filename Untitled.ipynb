```python
import queue

def bfs(point_list, r=25):
    depth = [0] * len(point_list)
    q = queue.Queue()
    
    q.put(0)
    depth[0] = 1
#     for i in range():
    while q.qsize():
        u = q.get()
        
        for i in range(len(point_list)):
            if distance(point_list[u], point_list[i]) <= 2*r and depth[i] == 0:
                depth[i] = depth[u] + 1
                q.put(i)
                
#     print(max(depth))
    return max(depth)

def num_sns(bs, sensors, r=25):
    count = 0
    for sensor in sensors:
        if distance(bs, sensor) <= 2*r:
            count += 1
    return count

radiues = []
for fn in sorted(os.listdir('data/small/hop')):
    if fn.startswith('.'):
        continue
    if not ('ga' in fn and 'r25' in fn and '_0' in fn):
        continue
    
    inp = WusnInput.from_file(f'data/small/hop/{fn}')
    points = [inp.BS, *inp.relays, *inp.sensors]
    
    radiues.append(bfs(points, r=inp.radius))
    print(fn, bfs(points, r=inp.radius), num_sns(inp.BS, inp.sensors))
    
#     print(len(x))
#     if 'r25' in fn:
#         print(fn)
#         print(np.mean([distance(tmp, inp.BS) for tmp in inp.sensors]))
#     break
#         print(max(x) - min(x), max(y) - min(y))
```
